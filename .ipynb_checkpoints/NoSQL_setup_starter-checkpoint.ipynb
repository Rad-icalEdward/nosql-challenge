## Part 2: Update the Database

1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
new_restaurant_add = {
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
establishments.insert_one(new_restaurant_add)
```

```python
# Check that the new restaurant was inserted
query = {"BusinessName":"Penang Flavours"}
results = establishments.find(query)
for business in results:
    pprint(business)
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
id_query = {"BusinessType":"Restaurant/Cafe/Canteen"}
fields = {"BusinessType", "BusinessTypeID"}
limit = 5

# Cast the results as a list and save them to a variable
results = list(establishments.find(id_query, fields).limit(limit))

# Pretty print the results
pprint(results)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
query = {"BusinessName":"Penang Flavours"}

update_value = {"$set": { "BusinessTypeID": 1}}

establishments.update_one(query, update_value)
```

```python
# Confirm that the new restaurant was updated
update_fields = {"_id", "BusinessName", "BusinessType", "BusinessTypeID"}

new_results = establishments.find(query, update_fields)
for business in new_results:
    pprint(business)
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
la_query = {"LocalAuthorityName":"Dover"}

la_results = establishments.find(la_query)
la_doc_count = la_results.count(True)
print(la_doc_count) 
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
la_delete = establishments.delete_many(la_query)
```

```python
# Check if any remaining documents include Dover
print(la_delete.deleted_count, " documents deleted.")
```

```python
# Check if any remaining documents include Dover
la_query = {"LocalAuthorityName":"Dover"}

la_results = establishments.find(la_query)
la_doc_count = la_results.count(True)
print(la_doc_count)
```

5. Some of the number values are stored as strings, when they should be stored as numbers. Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for latitude
establishments.update_many({}, [{'$set': {"geocode.longitude": {'$toDecimal': "$geocode.longitude"} 
                                         }}
                               ])
```

```python
# Change the data type from String to Decimal for longitude
establishments.update_many({}, [{'$set': {"geocode.latitude": {'$toDecimal': "$geocode.latitude"}
                                         }}
                                 ])
```

```python
# Check that the coordinates are now numbers
[business for business in establishments.find({}, ["geocode.longitude", "geocode.latitude"]).limit(limit)]
```

```python
# Change the data type from String to Integer for RatingValue to integer numbers
# I didn't think I was able to do this because there was mixed data with some integers and some strings so I did this workaround
# Find all the documents with a string type for the RatingValue
match_query = {'$match': {'RatingValue': {'$type' : 2}}}
sort_values = {'$sort': { 'count': -1 }}

# Put the pipeline together
pipeline = [match_query, sort_values]


results = list(establishments.aggregate(pipeline))

pprint(results[0:10])
```

```python
# Show just the relevant information for the documents before updating
[business for business in establishments.find({}, ["BusinessName", "RatingValue"]).limit(limit)]
```

```python
# Update the data type for RatingValue of documents with string data type to integer type
for business in results:
    business.update({'$set': {'RatingValue': {'$toInt': '$RatingValue'}
                                         }})
```

```python
# Check that the ratings are now numbers
# Show just the relevant information for the documents after updating to show they are now integers
[business for business in establishments.find({}, ["BusinessName", "RatingValue"]).limit(limit)]
```
