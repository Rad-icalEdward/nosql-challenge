### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {"scores.Hygiene":20}

hygiene_results = list(establishments.find(hygiene_query))

# Pretty print the results
pprint(hygiene_results)
```

```python
# Use count_documents to display the number of documents in the result
establishments.count_documents(hygiene_query)
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(hygiene_query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.json_normalize(hygiene_results)

# Display the number of rows in the DataFrame
print(f"The number of rows in this data frame is {len(hygiene_df)}")

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# Hint: The London Local Authority has a longer name than "London" so you will need to use $regex as part of your search.
london_rating_query = {"LocalAuthorityName": {'$regex': "London"}, "RatingValue": {"$gte":4}}

london_rating_results = list(establishments.find(london_rating_query))

# Pretty print the results
pprint(london_rating_results)
```

```python
# Use count_documents to display the number of documents in the result
establishments.count_documents(london_rating_query)
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(london_rating_query))
```

```python
# Convert the result to a Pandas DataFrame
london_rating_df = pd.json_normalize(london_rating_results)

# Display the number of rows in the DataFrame
print(f"The number of rows in this data frame is {len(london_rating_df)}")

# Display the first 10 rows of the DataFrame
london_rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Find latitude and longitude for Penang Flavours
pprint(establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude', 'geocode.longitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
penang_latitude = 51.4901
penang_longitude = 0.0838

penang_query = {"RatingValue": 5,
                "geocode.latitude": {"$lte": (penang_latitude + degree_search), "$gte": (penang_latitude - degree_search)}, 
                "geocode.longitude": {"$lte": (penang_longitude + degree_search), "$gte": (penang_longitude - degree_search)}
               } 
penang_sort =  [("scores.hygiene", 1)]

penang_results = list(establishments.find(penang_query).sort(penang_sort).limit(5))

# Pretty print the results
pprint(penang_results)
```

```python
# Convert the result to a Pandas DataFrame
penang_results_df = pd.json_normalize(penang_results)

penang_results_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Hint: You will need to use the aggregation method to answer this.
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {"scores.Hygiene": 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {"_id": "$LocalAuthorityName", 
                          "count": {"$sum": 1}}
              }

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {"scores.hygiene": -1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Print the number of documents in the result
local_zero_results = list(establishments.aggregate(pipeline))

print("Number of rows in result: ", len(local_zero_results))
```

```python
# Print the first 10 results
pprint(local_zero_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_zero_df = pd.json_normalize(local_zero_results)

# Display the number of rows in the DataFrame
print(f"The number of rows in this data frame is {len(local_zero_df)}")

# Display the first 10 rows of the DataFrame
local_zero_df.head(10)
```
